## Crime Data

```python
# Concatinate all crime data
concat_crime_df = pd.concat([df_2015, df_2016, df_2017, df_2018])
concat_crime_df.head()
```

```python
# Delete extraneous column
del concat_crime_df['Address']
del concat_crime_df['CaseNumber']
del concat_crime_df['CrimeAgainst']
del concat_crime_df['OccurTime']
del concat_crime_df['OffenseCategory']
del concat_crime_df['OpenDataLat']
del concat_crime_df['OpenDataLon']
del concat_crime_df['OpenDataX']
del concat_crime_df['OpenDataY']
del concat_crime_df['ReportDate']
del concat_crime_df['OffenseCount']
concat_crime_df.head()
```

```python
# Rename date field
concat_crime_df = concat_crime_df.rename(columns={"OccurDate":"Date"})
concat_crime_df.head()
```

```python
# Add day of week
# concat_crime['dow']= pd.to_datetime(concat_crime['OccurDate']).dt.dayofweek
# concat_crime.head()
```

```python
# Filter out reports of crimes prior to 2015
concat_crime_df['Date'] = pd.to_datetime(concat_crime_df['Date'])
start_date = '01/01/2015'
end_date = '12/31/2018'
mask = (concat_crime_df['Date'] > start_date) & (concat_crime_df['Date'] <= end_date)
concat_crime_df = concat_crime_df.loc[mask]
concat_crime_df
```

```python
crime_rates_df = concat_crime_df['Date'].value_counts()
crime_rates_df['Crimes'] = concat_crime_df.groupby(['Date'])['Neighborhood'].transform(len)
crime_rates_df.head()
```
