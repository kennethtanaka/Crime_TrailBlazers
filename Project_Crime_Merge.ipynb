```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Name of the CSV file
file15 = 'Resources/2015_Portland_Crime_Data.csv'
file16 = 'Resources/2016_Portland_Crime_Data.csv'
file17 = 'Resources/2017_Portland_Crime_Data.csv'
file18 = 'Resources/2018_Portland_Crime_Data.csv'
file19 = 'Resources/2019YTD_Portland_Crime_Data.csv'
nba_data ='Resources/portland.games.stats.csv'
```

```python
# The correct encoding must be used to read the CSV in pandas
df_2015 = pd.read_csv(file15, encoding="ISO-8859-1")
df_2016 = pd.read_csv(file16, encoding="ISO-8859-1")
df_2017 = pd.read_csv(file17, encoding="ISO-8859-1")
df_2018 = pd.read_csv(file18, encoding="ISO-8859-1")
df_2019 = pd.read_csv(file19, encoding="ISO-8859-1")
nba_data = pd.read_csv(nba_data, encoding="ISO-8859-1")
```

# Portland Trailblazers Stats 2015-19

```python
nba_data.head()
```

```python
#Created a new dataframe using the data_df to display columns these columns:
organized_nba_df = nba_data[['Team','Game','Date','Home',"WINorLOSS"]] 

organized_nba_df.head(10)
```

```python
#extract POR team data and reset index
trailblazers_df = organized_nba_df.loc[organized_nba_df["Team"] == "POR"]
new_trailblazers_df = trailblazers_df.reset_index()
#new_trailblazers_df.columns
new_trailblazers_df.head()
```

```python
#remove index column 
blazers_df = new_trailblazers_df[['Team', 'Game', 'Date', 'Home', 'WINorLOSS']]
blazers_df.head()
```

```python
#all home games regardless of wins and losses
home_blazers_df = blazers_df.loc[blazers_df["Home"] == "Home"]
home_blazers_df
```

```python
reset_home_blazers_df = home_blazers_df.reset_index()
reset_home_blazers_df
```

```python
reset_home_blazers_df['WINorLOSS'].value_counts()
```

```python
#reset_home_blazers_df.Date.values
reset_home_blazers_df['Date']
```

```python
x_df= pd.DataFrame(reset_home_blazers_df.Date.values)
x_df.columns = ['Date']
x_df.head(10)
```

```python
x_df['date_2']= pd.to_datetime(x_df['Date']).dt.date
x_df.head()
```

```python
x_df['dow']= pd.to_datetime(x_df['date_2']).dt.dayofweek
x_df.head(15)
```

```python
# all the wins at home per day, 0 is Monday
x_df["dow"].value_counts()
```

```python
#convert the 'OccurDate' column to datetime format 
df_2015['Date']= pd.to_datetime(df_2015['OccurDate'])

# Delete OccurDate after converting to datetime
del df_2015['OccurDate']

# create new column for day of week where Sunday=0, Monday=1, Tuesday=2, Wednesday=3, Thursday=4, Friday=5, Saturday=7, 
df_2015['DOW']= pd.to_datetime(df_2015['Date']).dt.dayofweek

#filter out dates that are not year 2016
df_2015 = df_2015[(df_2015['Date'].dt.year == 2015)]

len(df_2015)
```

```python
df_2015.head()
```

```python
#create dataframe of date and day of week
df_dow2015 = df_2015[["Date", "DOW"]]
df_dow2015.head()
```
