```python
import pandas as pd
import numpy as np
```

```python
nba_data = pd.read_csv("nba.games.stats.csv")
nba_data
```

```python
nba_data.head()
```

```python
#Created a new dataframe using the data_df to display columns these columns:
organized_nba_df = nba_data[['Team','Game','Date','Home',"WINorLOSS"]] 

organized_nba_df.head(10)
```

```python
#extract POR team data and reset index
trailblazers_df = organized_nba_df.loc[organized_nba_df["Team"] == "POR"]
new_trailblazers_df = trailblazers_df.reset_index()
#new_trailblazers_df.columns
new_trailblazers_df.head()
```

```python
#remove index column 
blazers_df = new_trailblazers_df[['Team', 'Game', 'Date', 'Home', 'WINorLOSS']]
blazers_df.head()
```

```python
#all home games regardless of wins and losses
home_blazers_df = blazers_df.loc[blazers_df["Home"] == "Home"]
home_blazers_df
```

```python
reset_home_blazers_df = home_blazers_df.reset_index()
reset_home_blazers_df
```

```python
reset_home_blazers_df['WINorLOSS'].value_counts()
```

```python
reset_home_blazers_df['Date']
```

```python
x_df= pd.DataFrame(reset_home_blazers_df.Date.values)
x_df.columns = ['Date']
x_df.head(10)
```

```python
x_df['date_2']= pd.to_datetime(x_df['Date']).dt.date
x_df.head()
```

```python
x_df['dow']= pd.to_datetime(x_df['date_2']).dt.dayofweek
x_df.head(15)
```

```python
# all the wins at home per day, 0 is Monday
x_df["dow"].value_counts()
```
